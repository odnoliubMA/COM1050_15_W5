```python
a , b = map(int, input().split())

def Swap( a,  b):
    tmp = b;
    b = a;
    a = tmp;
    return ( a , b);

print(Swap( a , b));
```

### W5A3

```python
from math import *


def prime(n):
        
    if (not ( n & 1) and n != 2 ) or n < 2:
        return "No";
    
    for i in range( 3 , int(sqrt(n)) + 1 , 2):
        if ( not ( n % i) ) : return "No"
    
    return "Yes";

n = int(input());

print(prime(n));
```

### W5A4

```python
def divisor(n):
    res = 1

    for i in range( 2 , int(sqrt(n) + 1)):
        if( not ( n % i)):
            res += i
            if ( n // i > i):
                res += ( n //i);

    return res == n;

n = int(input());

print( ("Yes" if divisor(n) else "No"))
```

```python
ls = list( map( int , input().split()));

k = int( input());

def brute( ls , k):
    for i in range(len(ls)):
        if( ls[i] == k):
            return i
    return -1;

print(brute( ls , k));
```

### W5A6

```python
n = int( input())

def fact( n):
    if n == 1: return 1;
    return n * fact( n - 1);

print(fact(n));
```

### W5A7

```python
a , b, c = map( str , input().split())

def calc( a, b ,c):
    _a = int(a)
    _b = int(b)

    if( c =='+'): return _a + _b
    elif( c =='-') : return _a - _b
    elif( c == '*'): return _a * _b
    return _a / _b

print(calc(a , b, c));
```

```python

def getBit(x , pos):
    return ( x >> pos) & 1;

cnt = 0;
a , b = map( int , input().split())


for i in range(64):
    _a = getBit( a , i);
    _b = getBit( b , i);

    cnt += ( _a != _b);

print( cnt);

```

### W5A9

```python
n = int(input())

def digit(n):
    res = 0;

    while n:
        res += n % 10;
        n //= 10

    return res;

print(digit(n))
```
